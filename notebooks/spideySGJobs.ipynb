```python
import matplotlib.pyplot as plt

def plot_matplotlib(pivot, categories=None, figsize=(12,6)):
    data = pivot[categories] if categories else pivot
    ax = data.plot(figsize=figsize, marker='o')
    ax.set_title('Category trends')
    ax.set_xlabel('Time')
    ax.set_ylabel('Value')
    ax.grid(True)
    plt.legend(title='Category')
    plt.tight_layout()
    plt.show()

```
